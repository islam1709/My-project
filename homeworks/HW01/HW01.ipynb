```python
# Импорт NumPy с общепринятым сокращением
import numpy as np
```

```python
# 1) Создание массива из списка
a = np.array([10, 20, 30, 40])
print("a:", a)

# 2) Создание через arange
b = np.arange(5, 25, 5)   # [5, 10, 15, 20]
print("b:", b)

# 3) Массивы zeros и ones
z = np.zeros(5)
o = np.ones((3, 2))
print("z:", z)
print("o:\n", o)
```

```python
# Двумерный массив (матрица)
matrix = np.array([
    [2, 4, 6],
    [8, 10, 12],
    [14, 16, 18]
])
print("matrix:\n", matrix)
```

```python
# Форма и тип данных
print("matrix.shape:", matrix.shape)
print("matrix.dtype:", matrix.dtype)
```

```python
# Доступ к элементам и срезам
element = matrix[1, 2]
print("matrix[1,2]:", element)

row = matrix[0]
print("Первая строка:", row)

column = matrix[:, 1]
print("Второй столбец:", column)

sub_matrix = matrix[:2, :2]
print("Подмассив:\n", sub_matrix)
```

```python
# Поэлементные операции
x = np.array([3, 6, 9])
y = np.array([2, 4, 6])

sum_xy = x + y
prod_xy = x * y
print("x + y =", sum_xy)
print("x * y =", prod_xy)
```

```python
# Статистические вычисления
print("x.mean() =", x.mean())
print("y.mean() =", y.mean())
print("matrix.sum() =", matrix.sum())
print("Среднее по строкам matrix:", matrix.mean(axis=1))
print("Среднее по столбцам matrix:", matrix.mean(axis=0))
```
